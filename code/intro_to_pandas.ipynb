2. column name matches a pandas DataFrame method

3. column name matches a restricted python word. See the link for the [full list](https://docs.python.org/2.5/ref/keywords.html)

#### Multiple Conditions

You can also pass more than one condition to the selection brackets. Each condition gets wrapped in parentheses if you have more than one. Let's consider three examples of groups that we might want to search for:
1. Employees in the sales department with a job satisfaction greater than or equal to three 
2. Employees in either Sales or Human Resources Departments
3. Employees in Sales who travel frequently

```python
# returns employees in Sales with a JobSatisfaction score of 3 or more
(employees.Department == 'Sales') & (employees.JobSatisfaction >= 3)

# There are three departments in our dataset so there are two ways to get employees in Sales or HR
# Method 1
# use the bitwise or operator
# returns employees in either the Sales Department or the Human Resources Department
(employees.Department == 'Sales') | (employees.Department == 'Human Resources')
                                     
# Method 2
# use the bitwise not operator
# returns employees not in Research & Development which is by default those in HR or Sales
~(employees.Department == 'Research & Development')                                   


# returns employees in Sales who travel frequently
(employees.Department == 'Sales') & (employees.BusinessTravel == 'Travel_Frequently')

```

Any of these results can be used to create a new DataFrame that you can use for additional analysis

``` python
# create a DataFrame of Sales employees who are frequent travelers and store it in a new variable
# note the use of the copy() function. without this changes made to the new DataFrame will also be reflected on the original
frequent_travelers = employees[(employees.Department == 'Sales') & (employees.BusinessTravel == 'Travel_Frequently')].copy()
```

### Renaming Columns and Indices

Pandas DataFrames have a [rename method](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rename.html) that can be used to rename columns and indices. 

The third column in our dataset is called BusinessTravel. We can use the rename method and a Python dictionary to change the name to 'frequent_traveler'

##### Python Dictionary
A Python dictionary is made up of key-value pairs wrapped in curly braces. It can be used to map values. Here we want to map the original column name BusinessTravel to the new column name frequent_traveler. In our example, we're only updating one column name, but your dictionary can contain more than one as long as each pair is separated by a comma.

**Note**: If you include inplace = True as a parameter to the method you'll modify the original employees DataFrame otherwise you won't see your changes reflected. I typically don't use inplace operations and pass my changes to a new DataFrame

```python

{'BusinessTravel':'frequent_traveler'} 

#renames the BusinessTravel column and modifies the original DataFrame
```

Duplicate values can make their way into datasets through both human or systems error. Let's use the example DataFrame below that has a list of employees, their names, and date of PTO taken. There are a few records with duplicate IDs and one where the entire row is duplicated. Pandas makes it easy to find duplicate values using the .duplicated() method. 

|    |   id | name          | pto_date        |
|---:|-----:|:--------------|:--------------------|
|  0 |    1 | Johan James   | 2021-01-02 00:00:00 |
|  1 |    1 | Johan James   | 2021-01-15 00:00:00 |
|  2 |    2 | Patty Smythe  | 2021-03-05 00:00:00 |
|  3 |    2 | Patty Smythe  | 2021-04-11 00:00:00 |
|  4 |    3 | Amanda Apple  | 2021-07-09 00:00:00 |
|  5 |    4 | Arnold Jones  | 2021-10-11 00:00:00 |
|  6 |    5 | Paul Nicholas | 2021-11-15 00:00:00 |
|  7 |    6 | Ira Albert    | 2021-12-01 00:00:00 |
|  8 |    6 | Ira Albert    | 2021-12-01 00:00:00 |


#### Finding Duplicates

``` python

# using keep = False returns all records
# can use keep = 'first' or 'last' to return either the first or last record

#passing this to the bracket notation returns the details for those rows
```

````python
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(11,10),constrained_layout=True)
gs = GridSpec(3, 2, figure=fig)
ax1 = fig.add_subplot(gs[0, 0])


df = (pd.crosstab(employees.department, employees.attrition, margins=True).reindex(['Yes','No'], 
                                                               axis = 1).sort_values(by='No'))

df.rename(index = {'Human Resources':'HR', 'Research & Development': 'R&D'}, inplace = True)


df.plot.barh(stacked = True, color = {'Yes':'teal','No':'lightgray'}, ax = ax1)
ax1.set_title('Attrition by Department', loc = 'left', pad = 35, size = 18)

ax1.set_ylabel('')
ax1.legend(ncol = 2, frameon = False, loc = 'upper left', bbox_to_anchor = (-0.03,1.08))


for spine in ['top', 'right']:
    ax1.spines[spine].set_visible(False)
    
    
for spine in ['bottom', 'left']:
    ax1.spines[spine].set_position(('outward', 10))    
    
ax1.tick_params(bottom = False, left = False, labelsize = 12)    

for p in ax1.containers[-1].patches:
    x, y = p.get_xy()
    width = p.get_width()
    height = p.get_height()
    attr_rate = x/(x+width)
    
    ax1.annotate(f'{attr_rate:0.0%}', xy = (x,y+height/2), va = 'center', size = 12)
    
    
fig.text(x = 0.06, y = 0.92, s='The Sales Organization is Driving Attrition ', size = 12)    



ax2 = fig.add_subplot(gs[0, 1])

tab = sm.stats.Table(pd.crosstab(employees.department, employees.attrition))
print(tab.test_nominal_association())


obs_vs_exp = pd.concat([tab.fittedvalues.Yes, tab.table_orig.Yes], axis = 1)
obs_vs_exp.columns = ['Expected', 'Actual']
obs_vs_exp.rename(index = {'Human Resources':'HR', 'Research & Development': 'R&D'}, inplace = True)
obs_vs_exp = obs_vs_exp.sort_values(by = 'Actual')


obs_vs_exp.plot.barh(color = {'Actual':'teal','Expected':'paleturquoise'}, ax = ax2)
ax2.set_title('Attrition: Actual vs. Expected', loc = 'left', pad = 35, size = 18)

ax2.set_ylabel('')
ax2.legend(ncol = 2, frameon = False, loc = 'upper left', bbox_to_anchor = (-0.03,1.09))


for spine in ['top', 'right']:
    ax2.spines[spine].set_visible(False)
    
    
for spine in ['bottom', 'left']:
    ax2.spines[spine].set_position(('outward', 10))    
    
ax2.tick_params(bottom = False, left = False, labelsize = 12)    

# adds annotation to right bar    
ax2.annotate('Sales Had 20 More \nExits Than Expected', 
            xy = (82, 0.95), xytext = (.315,.24), textcoords = 'axes fraction', size = 12,
            arrowprops = dict(arrowstyle='-[, widthB=1.5, lengthB=0.5', lw=1))    


ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])
ax5 = fig.add_subplot(gs[2, 0])
ax6 = fig.add_subplot(gs[2, 1])

fig.suptitle("GridSpec")

````
